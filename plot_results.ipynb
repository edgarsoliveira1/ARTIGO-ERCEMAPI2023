```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv('./results.csv')

df = pd.DataFrame(data)
df
```

```python
sns.set(font_scale=2)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))
# fig.suptitle('Comparação de Métricas por Método')
sns.barplot(df,x='Método', y='Acurácia', ax=axes[0, 0])
axes[0, 0].set_title('Acurácia')
axes[0, 0].set_xlabel('')
axes[0, 0].set_ylabel('')
sns.barplot(df,x='Método', y='Precisão', ax=axes[0, 1])
axes[0, 1].set_title('Precisão')
axes[0, 1].set_xlabel('')
axes[0, 1].set_ylabel('')
sns.barplot(df,x='Método', y='Recall', ax=axes[1, 0])
axes[1, 0].set_title('Recall')
axes[1, 0].set_xlabel('')
axes[1, 0].set_ylabel('')
sns.barplot(df,x='Método', y='F1', ax=axes[1, 1])
axes[1, 1].set_title('F1-score')
axes[1, 1].set_xlabel('')
axes[1, 1].set_ylabel('')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
